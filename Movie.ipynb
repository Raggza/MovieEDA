## 1. Import thư viện và dataset

```python
#Import thư viện cần thiết

import pandas as pd
import numpy as np
```

```python
#Đọc file csv

df1 = pd.read_csv('movies.csv')
df1
```

## 2. Kiểm tra thông tin cơ bản của dataset

```python
#Kiểm tra kiểu dữ liệu các cột

df1.dtypes
```

```python
#Kiểm tra số hàng và số cột

df1.shape
```

```python
#Kiểm tra dữ liệu bị lặp

df1[df1.duplicated()]
```

```python
#Kiểm tra số lượng dữ liệu bị thiếu ở các cột (số lượng và % so với tổng dữ liệu)

for col in df1.columns:
    no_missing = df1[col].isnull().sum()
    pct_missing = (no_missing / df1.shape[0]) * 100
    print('{} - {} - {:.2f}%'.format(col, no_missing, pct_missing))
```

## 3. Làm sạch dữ liệu

```python
#Tạo một bản copy để thực hiện làm sạch

m01 = df1.copy()
```

```python
#Loại bỏ những dòng có giá trị null ở các cột rating, released, writer, star, country, company, runtime

m01.dropna(subset=['rating', 'released', 'writer', 'star', 'country', 'company', 'runtime'], inplace = True)

pct_drop = ((df1.shape[0] - m01.shape[0]) / df1.shape[0]) * 100
print('{:.2f}%'.format(pct_drop))
```

```python
m01.shape
```

```python
#Thay thế giá trị null thành 0 ở các cột score, vote, budget, gross

m01['score'].fillna(0, inplace = True)
m01['votes'].fillna(0, inplace = True)
m01['budget'].fillna(0, inplace = True)
m01['gross'].fillna(0, inplace = True)

m01.shape
```

```python
#Tạo một bản copy để làm sạch

m02 = m01.copy()
```

```python
def get_date(date):
    if '(' in date:
        date = date[:date.index('(')]
    return date[:-1].strip()

m02['released'] = m02.released.apply(lambda date: get_date(str(date)))

m02["released"]
```

```python
m02["released"] = m02["released"].astype('datetime64[ns]')
```

```python
m02
```

## 4. Xuất file csv

```python
#Tạo file csv mới sau khi làm sạch dữ liệu

m02.to_csv('movies_clear.csv', index=False)
```
